# Step 9: Pie Chart Representation of Sub Categories demand

```python
plt.figure(figsize=(12,12))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(7,12))
df['Region'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

### OBSERVATION 5: Sales are higher in Western region

```python
plt.figure(figsize=(20,6))
sns.countplot(x='State',data=df,palette='rainbow',order=df['State'].value_counts().index)
plt.xticks(rotation=75)
plt.show()
```

### OBSERVATION 6: California has the highest sale, followed by New York and Texas

```python
df.hist(figsize=(10,10))
```

### OBSERVATION 7: Mostly people buy 2 quantity of products and mostly discount varies from 0-20%

# Step 10: Line Plot Graph Representation

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

### OBSERVATION 8: Beyond 20% discount will result in Loss

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

### OBSERVATION 9: Maximum profit when quantity sold is in between 12-14

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar()
plt.ylabel('Profit and sales')
plt.show()
```

### OBSERVATION 10: Profit and Sales are maximum in the Consumer Segment and Minimum in Home Office Segment

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='PuRd_r')
plt.xlabel('Region',fontsize=15)
plt.show()
```

### OBSERVATION 11: Segment wise sales in each Region is almost same

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','green'],figsize=(8,5))
plt.ylabel('Profit and sales')
plt.show()
```

### OBSERVATION 12: Profit and Sales are maximum in western region and minimum in southern region

```python
plt.figure(figsize=(20,8))
plt.title("Region and Quantity Ordered")
sns.countplot("Region",data=df,hue="Quantity",color='black')
plt.show()
```

### OBSERVATION 13: Western and Eastern regions order more quantity

```python
df.groupby('State')[['Profit','Sales']].sum().plot.bar(color=['purple','pink'],figsize=(20,8))
plt.ylabel('Profit and Sales  across States')
plt.show()
```

### OBSERVATION 14: California and New York yield most profit, while Texas, Ohio and Pennsylvania yield most loss

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['orange','yellow'],figsize=(8,5))
plt.ylabel('Profit and sales ')
plt.show()
```

### OBSERVATION 15: Technology and Office Supplies categories yield higher profit compared to Furniture category

```python
df.groupby('Sub-Category')[['Profit','Sales']].sum().plot.bar(color=['blue','cyan'],figsize=(20,8))
plt.ylabel('Profit and Sales')
plt.show()
```

### OBSERVATION 16: Phones and Chairs have same number of sales but Phones yields more profit than Chairs. Whereas Tables yield most loss.

```python
df_region = df.groupby(['Region'])[['Sales', 'Discount', 'Profit']].mean()
df_region
```

```python
df_region.plot.pie(subplots=True, figsize=(22, 25), autopct='%1.1f%%', labels = df_region.index)
plt.title('Pie Chart based on Region',fontsize=20)
plt.show()
```

### OBSERAVATION 17 : Maximum Sales in South, Minimum Sales in Central. Maximum Discount in Central, Minimum Discount in West. Maximum Profit in West, Minimum Profit in Central.
