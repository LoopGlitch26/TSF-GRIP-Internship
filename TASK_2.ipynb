# Step 7: K-Means Clustering

Finding optimal number of clusters using the elbow method 

```python
from sklearn.cluster import KMeans
```

```python
x = df.iloc[:,[0,1,2,3]].values     
x
```

Initializing the list for the values of WCSS

```python
WCSS = []
# Within Clusters Sum of Squares
```

Initialized the for loop for the iteration on a different value of k ranging from 1 to 10; since for loop in Python excludes the outbound limit, so it is taken till 11 to include 10th value

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df)
    WCSS.append(kmeans.inertia_)
```

# Step 8: Plotting Line Graph for Elbow Method

```python
plt.plot(range(1,11),WCSS, color='Orange', marker='o', markerfacecolor='Red')
plt.title("The Elbow Method Graph")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
```

# Step 9: Training the K-Means Model on Iris Dataset

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_predict = kmeans.fit_predict(x)
print("TRAINING COMPLETED")
y_predict
```

# Step 10: Scatter Plot Graph Visualiation for Clusters and Centroids

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(15,15))

ax1.scatter(df['sepal length (cm)'], df['sepal width (cm)'], s=50, edgecolor="Blue", linewidth=1)
ax1.set_title('Before Clustering')
ax1.set(xlabel='Sepal Length in cm', ylabel='Sepal Width in cm')

ax2.scatter(x[y_predict==0,0], x[y_predict==0,1], s=50, color='Lightgreen', edgecolor="Green", linewidth=1, label='Iris Setosa')
ax2.scatter(x[y_predict==1,0], x[y_predict==1,1], s=50, color='Cyan', edgecolor="Blue", linewidth=1, label='Iris Versicolor')
ax2.scatter(x[y_predict==2,0], x[y_predict==2,1], s=50, color='Violet', edgecolor="Purple", linewidth=1, label='Iris Virginica')
ax2.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, color='Yellow', edgecolor="Red", linewidth=1, label="Centroids")
ax2.legend(loc='best')
ax2.set_title('After Clustering')
ax2.set(xlabel='Sepal Length in cm', ylabel='Sepal Width in cm')

ax3.scatter(df['petal length (cm)'], df['petal width (cm)'], s=50, edgecolor="Blue", linewidth=1)
ax3.set_title('Before Clustering')
ax3.set(xlabel='Petal Length in cm', ylabel='Petal Width in cm')

ax4.scatter(x[y_predict==0,2], x[y_predict==0,3], s=50, color='Lightgreen', edgecolor="Green", linewidth=1, label='Iris Setosa')
ax4.scatter(x[y_predict==1,2], x[y_predict==1,3], s=50, color='Cyan', edgecolor="Blue", linewidth=1, label='Iris Versicolor')
ax4.scatter(x[y_predict==2,2], x[y_predict==2,3], s=50, color='Violet', edgecolor="Purple", linewidth=1, label='Iris Virginica')
ax4.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, color='Yellow', edgecolor="Red", linewidth=1, label="Centroids")
ax4.legend(loc='best')
ax4.set_title('After Clustering')
ax4.set(xlabel='Petal Length in cm', ylabel='Petal Width in cm')
```
